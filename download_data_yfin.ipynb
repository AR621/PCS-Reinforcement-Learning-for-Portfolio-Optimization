```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["AAPL", "AMZN", "GOOG", "MSFT", "NVDA", "^GSPC"]
start_date = '1994-01-01'
end_date = '2024-03-31'
```

```python
data = {}
for ticker in tickers:
    data[ticker] = yf.download(ticker, start=start_date, end=end_date, interval="1d")
```

```python
# save to csv
timestamp = pd.Timestamp.now().strftime("%Y-%m-%d-%H-%M")
# file_name = f'{ticker}_{timestamp}.csv'
for ticker in tickers:
    file_name = f'{ticker}.csv'
    file_path = f'./data/stocks/{file_name}'
    data[ticker].to_csv(file_path)
```
